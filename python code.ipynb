```python
df['Price'] = df['Price'].round(2)
df['Costs'] = df['Costs'].round(2)
```

```python
df['Product type'].value_counts().head()
```

```python
df['Product type'].value_counts().head().plot(kind= 'line',color = plt.cm.tab20.colors)
plt.ylabel('count')
plt.xlabel('Product Type')
plt.show()
```

```python
total_revenue_by_product = df.groupby('Product type')['Revenue generated'].sum().reset_index().round(2)
print(total_revenue_by_product)
```
